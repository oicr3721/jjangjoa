댓글 창 한 페이지 분량만큼
반응 표시

```python
import requests
from bs4 import BeautifulSoup

# 스팀 이벤트에 대한 댓글 페이지
url = "https://steamcommunity.com/app/367520/eventcomments/807972774925588882/?ctp="

res = requests.get(url)
soup = BeautifulSoup(res.text, "html.parser")

comment_comment = soup.find_all("div", class_="commentthread_comment_text")
comment_name = soup.find_all("bdi")
comment_date = soup.find_all("span", class_="commentthread_comment_timestamp")

data1 = []
data2 = []
data3 = []

# 댓글 내용 리스트와 이름 리스트에 각각 저장
for n in comment_comment:
    data1.append(n.text.strip())
for m in comment_name:
    data2.append(m.text)
for o in comment_date:
    data3.append(o.text.strip())

print(data1)
print(data2)
print(data3)

data = {
    "댓글 내용": data1,
    "이름": data2,
    "날짜": data3
}
```

```python
# 데이터 프레임으로 변환
import pandas as pd

df = pd.DataFrame(data)
df.to_csv("스팀 반응.csv")
print(df)
```

```python
import requests
from bs4 import BeautifulSoup

# 스팀 이벤트에 대한 댓글 페이지
url = "https://steamcommunity.com/app/367520/eventcomments/807972774925588882/?ctp="
ctp = 1 # ctp = 댓글 페이지 수

res = requests.get(url+str(ctp))
soup = BeautifulSoup(res.text, "html.parser")

# 댓글 페이지 수 계산
comment_count = soup.find("div", class_="forum_paging_summary ellipsis")
comment_max = comment_count.text.strip()
comment_page = int(comment_max.split()[3])//15 + 1 # comment_max.split()[3] = 207은 문자열
# 한 페이지당 최대 댓글 수가 15개인 걸 이용하여 댓글 페이지 수를 계산, 댓글이 15개보다 적을 경우를 대비해 1을 더해줌-> ctp는 1부터 시작하기에

data1 = []
data2 = []
data3 = []

# 댓글 페이지 수 만큼 반복하여 리스트에 댓글 정보를 저장하는 반복문
for i in range(comment_page):

    comment_comment = soup.find_all("div", class_="commentthread_comment_text")
    comment_name = soup.find_all("bdi")
    comment_date = soup.find_all("span", class_="commentthread_comment_timestamp")

    # 댓글 내용 리스트와 이름 리스트에 각각 저장
    for n in comment_comment:
            data1.append(n.text.strip())
    for m in comment_name:
            data2.append(m.text)
    for o in comment_date:
            data3.append(o.text.strip())

    # 변동되는 ctp에 따라 다시 크롤링 하는 페이지 재정의
    ctp += 1
    res = requests.get(url+str(ctp))
    soup = BeautifulSoup(res.text, "html.parser")



print("데이터 개수:",len(data1))
print(data1)
print(data2)
print(data3)

data = {
    "댓글 내용": data1,
    "이름": data2,
    "날짜": data3
}
```

```python
# 데이터 프레임으로 변환하고 csv 파일로 저장까지
import pandas as pd

df = pd.DataFrame(data)
df.to_csv("스팀 반응.csv")
print(df)
```
