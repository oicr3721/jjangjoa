## 분위수(quartile)

데이터를 순서대로 늘어 놓았을 때 이를 균등한 간격으로 나누는 기준점

- 이분위수는 전체 데이터를 두 구간으로 나누기 때에 중앙값에 해당

- 사분위수는 3개가 나오고 각각 25%, 50%, 75%에 해당

- 25%에 대항하는 값을 제 1 사분위수, 중앙값을 제2사분위수, 75%에 해당하는 값을 제3사분위수라고도 부른다.
image.png

##**quantile()메서드**

```python
# 하위 25%에 위치한 값을 출력하려면 0.25를 입력한다.
ns_book7['대출건수'].quantile(0.25)
```

 qunatile()메서드에 여러 개의 분위수를 지정하면 각 분위수에 해당하는 값을 담은 시리즈 객체를 반환

```python
# 1,2,3,4,5 다섯 개의 숫자가 있을 때 90% 위치에 있는 값은?
# 시리즈 객체를 정의한 후 분위수를 구하기
pd.Series([1,2,3,4,5]).quantile(0.9)
```

####  interpolation매개변수
위 코드는 interpolatoin매개변수를 따로 지정하지 않아서 기본값으로 계산되었다.

예를 들어 4와 5사이에서 90%에 해당하는 분위수를 찾을때, 다음그림처럼 4의 분위수 0.75와 5의 분위수 1.0을 사용해 간단한 비례식을 세울 수 있다.

여기서 궁금한 값 b에 대해 정리하면 아래와 같은 식을 얻게 된다.
image.png

이렇게 두 지점 사이에 놓인 특정 위치의 값을 구하는 방법을

**보간(interploation)**이라고 부른다.

```python
# linear 방식
4 + (0.9-0.75)*(5-4)/(1.0-0.75)
```

```python
pd.Series([1,2,3,4,5]).quantile(0.9,interpolation='midpoint')
```

interpolation매개변수는 'linear'외에도 여러 가지 다른 보간 방식을 제공

'midpoint'는 분위수에 상관없이 무조건 두 수 사이의 중앙값을 사용한다.

```python
#nearest'는 두 수 중에서 가까운 값을 선택한다. 아래의 경우 0.9가 1.0에 더 가깝기 때문에 5가 된다.
pd.Series([1,2,3,4,5]).quantile(0.9,interpolation='nearest')
```

## 백분위구하기

1,2,3,4,5에서 4의 백분위는 0.75라고 쉽게 찾을 수 있다.

데이터가 작을 때는 이렇게 직접 쉽게 헤아려 구할 수 있지만 데이터가 많으면 어렵다.

예를들어 남산 도서관 대출 데이터에서 대출건수 10이 위치한 백분위를 어떻게 찾을 수 있을까?

판다스에서 이에 대한 메서드를 제공하지는 않지만, 간단한 트릭으로 유사한 결과를 얻을 수 있다.

먼저 '대출건수'열의 값이 10보다 작은지 비교하여 불리언배열을 만든다.
image.png

```python
borrow_10_flag = ns_book7['대출건수'] < 10
```

이렇게 변환한 borrow_10_flag배열(실제로는 시리즈 객체)에서 True인 개수를 모두 헤아린 후 전체 데이터 개수로 나누면 10보다 작은 대출 건수의 비율을 얻을 수 있다.

데이터의 양이 많다면 전체 데이터를 순서대로 놓았을 때 10이 위치한 백분위와 거의 비슷하게 된다.

판다스에서 불리언 자료를 산술 연산하면 True는 1, False는 0으로 취급한다.

따라서, 다음처럼 borrow_10_flag배열에서 mean()메서드를 호출하여 평균을 구하면 10보다 작은 값이 차지하는 비율을 간단히 얻을 수 있다.

```python
borrow_10_flag.mean()
```

10에 대한 백분위는 0.65정도임을 알 수 있다. quantile()메서드에 백분위를 넣어 직접확인해보자.

```python
ns_book7['대출건수'].quantile(0.65)
```

## 분산(variance)

분산은 평균으로부터 데이터가 얼마나 퍼져있는지를 나타내는 통계량이다.

데이터가 가운데 모여 있다면 분산이 작고 넓게 퍼져있다면 분산이 크다.
image.png

분산은 데이터의 각 값에서 평균을 뺀 다음 제곱한 후 평균처럼 샘플 개수로 나누어 구할 수 있다.
image.png

image.png

$ s^2 = \dfrac{\sum_{i=1}^{n}(x_i - \bar{x})^2}{n}$

$ \bar{x} = \dfrac{\sum_{i=1}^n x_i}{n}$

```python
ns_book7['대출건수'].var()
```

분산은 데이터가 평균에서 얼마나 퍼져있는지 나타내는 값이기 때문에 이 결과만 보고 값이 작은지 큰지 판단하기 어렵다.

분산은 제곱을 했기 때문에 데이터가 평균에서 멀어질수록 값이 급격히 커진다.

또 단위가 대출건수의 제곱이 되어 해석하기 어렵다.

이를 보완하기 위해 분산에 제곱근을 취한 표준편차를 사용한다.

## 표준편차 (standard deviation)

표준편차는 분산에 제곱근을 한 것으로 수식 기호는 s를 사용

분산은 표준편차를 제곱한 것이므로 분산을 s제곱으로 표시
image.png

$ s = \sqrt{\dfrac{\sum_{i=1}^{n}(x_i - \bar{x})^2}{n}}$

$\sqrt{4}=2$

```python
ns_book7['대출건수'].std()
```

도서대출 건수는 평균 11회

평균보다 훨씬 큰 대출건수가 있기 때문에 표준편차가 큰 것으로 예상할 수 있다.


## 수식을 코드로 구현

판다스와 넘파이를 사용해 표준편차를 수식대로 직접 계산해서 판다스의 std()메서드를 사용한 결과와 동일한지 확인하기
image.png

```python
import numpy as np

diff = ns_book7['대출건수'] - ns_book7['대출건수'].mean()

np.sqrt( np.sum(diff**2) / (len(ns_book7)-1) )
```
